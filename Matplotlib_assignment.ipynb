```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
df=pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
#df.head()
import numpy as np
df=df.sort(['Date'])
df['Year']=df['Date'].apply(lambda x:x[:4])
df['Month-Date']=df['Date'].apply(lambda x:x[5:])
df=df[df['Month-Date']!= '02-29']

df_temp_min=df[(df['Element']=='TMIN') & (df['Year']!='2015')]
df_temp_max=df[(df['Element']=='TMAX') & (df['Year']!='2015')]
temp_min=df_temp_min.groupby('Month-Date')['Data_Value'].agg({'tempminexcept2015mean':np.mean})
temp_max=df_temp_max.groupby('Month-Date')['Data_Value'].agg({'tempmaxexcept2015mean':np.mean})

#temp_min
#or use as_index=False
temp_max=temp_max.reset_index() #will have 364 index
temp_min=temp_min.reset_index()
#temp_max
plt.figure()
plt.plot(temp_min['tempminexcept2015mean'],alpha=0.75,color='r',label='Low Each Day')
plt.plot(temp_max['tempmaxexcept2015mean'],alpha=0.75,color='k',label='High Each Day')
#plt.show()
#for year 2015
df_temp_min15=df[(df['Element']=='TMIN') & (df['Year']=='2015')]
df_temp_max15=df[(df['Element']=='TMAX') & (df['Year']=='2015')]
temp_min_15=df_temp_min15.groupby('Month-Date')['Data_Value'].agg({'tempmin2015mean':np.mean})
temp_max_15=df_temp_max15.groupby('Month-Date')['Data_Value'].agg({'tempmax2015mean':np.mean})
temp_min_15=temp_min_15.reset_index()
temp_max_15=temp_max_15.reset_index()

#min_broken=temp_min_15[temp_min_15['tempmin2015mean']<temp_min['tempminexcept2015mean']].index.tolist()
#to track points of extreme index.tolist is used 
min_broken=temp_min_15[temp_min_15['tempmin2015mean']<temp_min['tempminexcept2015mean']].index.tolist()
max_broken=temp_max_15[temp_max_15['tempmax2015mean']>temp_max['tempmaxexcept2015mean']].index.tolist()
#max_broken
plt.scatter(min_broken,temp_min_15['tempmin2015mean'].iloc[min_broken],color='orange',label='Min Broken')
plt.scatter(max_broken,temp_max_15['tempmax2015mean'].iloc[max_broken],color='pink',label='Max Broken')
#plt.show()
plt.xlabel('Month')
plt.ylabel('Temperature in degree Celcius')
plt.title('Temperature from 2005-2014 and Extreme in 2015')
plt.gca().fill_between(range(len(temp_max)),temp_min['tempminexcept2015mean'],temp_max['tempmaxexcept2015mean'],color='yellow',alpha=0.5)
#plt.show()

a=[30,60,90,120,150,180,210,240,270,300,330,360]
b=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.xticks(a,b)
plt.legend(loc='best')
plt.show()
```
